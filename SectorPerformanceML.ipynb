```
# display columns and select what is needed

# ScikitLearn piipeline infrastructure

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import Imputer, LabelEncoder
from sklearn.preprocessing  import LabelBinarizer

from sklearn.base import TransformerMixin #gives fit_transform method for free
class MyLabelBinarizer(TransformerMixin):
    def __init__(self, *args, **kwargs):
        self.encoder = LabelBinarizer(*args, **kwargs)
    def fit(self, x, y=0):
        self.encoder.fit(x)
        return self
    def transform(self, x, y=0):
        return self.encoder.transform(x)
      
class DataFrameSelector(BaseEstimator, TransformerMixin):
  '''
    transform the data by selecting desired attributes,
    dropping the rest and converting the dataframe to
    a NumPy array
  '''
  def __init__(self, attribute_names):
    print ('Initializedwith :{}'.format(attribute_names ))
    self.attribute_names = attribute_names
    
  def fit(self, X, y=None):
    return self
  
  def transform(self, X):
    return X[self.attribute_names].values

  
def train_test_split(input_df):
    from sklearn.model_selection import train_test_split
    return train_test_split(input_df, test_size=0.2, random_state = 42)

def replace_invalid_chars(input_df, invalidChar):
  return input_df.replace(invalidChar, np.nan)

def make_pipeline(estimator_list):
  return Pipeline(estimator_list)

all_data = advanced_df1[valid_columns].fillna({'Sector': 'NA'})


num_attribs = [col for col in all_data.columns if col not in  ['Symbol','Sector', 'industry', 'advancing']]
cat_attribs = ['Sector']
print (num_attribs)
print (cat_attribs)

# Split data into Train and Test
train, test = train_test_split(all_data)

m_data_features = train.drop('advancing', axis=1)
m_data_labels = train['advancing'].copy() # Bear this in mind, we'll use it later

num_pipeline = make_pipeline([
  ('selector', DataFrameSelector(num_attribs)),
  ('imputer', Imputer(strategy='most_frequent')),
  ('std_scaler', StandardScaler()),
  ])

cat_pipeline = make_pipeline([
  ('selector', DataFrameSelector(cat_attribs)),
  ('lencoder', MyLabelBinarizer()),
])

from sklearn.pipeline import FeatureUnion
full_pipeline = FeatureUnion(transformer_list = [
  ("num_pipeline", num_pipeline),
  ("cat_pipeline", cat_pipeline)
])



train_data_prepared = full_pipeline.fit_transform(train)
#train_data_cat_prepared = cat_pipeline.fit_transform(m_data_features)
#print(sklearn.base.__version__)
```

<h2> Training and Evaluating on the Training Set</h2>

```
# now we run few ML algorithm to see how good we are
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline


def get_all_estimators():
  return [  make_pipeline([
                    ('svc1', SVC(kernel="linear", C=0.025))
                    ]),
            make_pipeline([
                      ('svc2',  SVC(gamma=2, C=1))
                    ]),
            make_pipeline([
                      ('decisionTreeClassifier', DecisionTreeClassifier())
                    ]),
            make_pipeline([ 
                      ('neuralNetwork',  MLPClassifier(alpha=1))
                    ]),
            make_pipeline([
                      ('rfClassifier', RandomForestClassifier())
                    ]),
            make_pipeline([
                      ('KNeigbors', KNeighborsClassifier())
                    ]),
            
        ]
```

<h2>Checking scores on Training Data for All Estimators </h2>

```
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_score, recall_score

def get_est_scores(estimator, x_train, y_train):
  return cross_val_score(estimator, x_train, y_train, scoring="neg_mean_squared_error", cv=10)

def get_precision_scores(estimator, x_train, y_train):
  y_train_pred = cross_val_predict(estimator, x_train, y_train)
  return precision_score(y_train, y_train_pred)

def get_f1_scores(estimator, x_train, y_train):
  from sklearn.metrics import f1_score
  y_train_pred = cross_val_predict(estimator, x_train, y_train)
  return f1_score(y_train, y_train_pred)


def evaluate(X, y, estimators) :
  eval_data = []
  scores = []
  for estimator in estimators:
    est_scores = get_est_scores(estimator, X, y)
    rmse_scores = np.sqrt(-est_scores)
    precision = get_precision_scores(estimator, X, y)
    
    eval_data.append((estimator.named_steps.keys()[0], precision))#rmse_scores.mean()))
    scores.append((estimator.named_steps.keys()[0], get_f1_scores(estimator, X, y)))
  return sorted(eval_data, key=lambda tpl: tpl[1]), sorted(scores, key=lambda tpl: tpl[1])
            
print ('MDataPrepared is:{}'.format(type(m_data_prepared)))
print ('Labels is:{}%'.format(type(m_data_labels)))
  
est_scores, f1_scores = evaluate(m_data_prepared, m_data_labels, get_all_estimators())

from pprint import pprint
pprint('---  Precision-----')
pprint(est_scores)
print('---f1 scores')
pprint(f1_scores)
```
